Question 3: Explain the difference between discrete and continuous distributions

1. Discrete Distribution

A discrete distribution is used when the random variable takes countable values.

**Characteristics:**

Values are finite or countably infinite (0,1,2,3‚Ä¶)

Probability is assigned to each specific value

Uses PMF (Probability Mass Function)

Total probability = 1

**Example:**

Number of heads in 3 coin tosses
Possible values: {0,1,2,3}

Common discrete distributions:

Binomial distribution

Poisson distribution

Geometric distribution

2. Continuous Distribution

A continuous distribution is used when the random variable can take infinite values within a range.

**Characteristics:**

Values lie in an interval (like 0 to 10)

Probability of a single value = 0

Uses PDF (Probability Density Function)

Probability is calculated over an interval

Total area under the curve = 1

**Example:**

Height of students (160.2 cm, 160.25 cm, etc.)

Common continuous distributions:

Normal distribution

Exponential distribution

Uniform distribution

Question 4: What is a binomial distribution, and how is it used in probability?


A binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes:

Success (with probability p)

Failure (with probability 1 ‚àí p)


It is used when:

1.The number of trials (n) is fixed

2.Each trial has only two outcomes (Yes/No, Pass/Fail, Head/Tail)

3. Trials are independent

4.Probability of success (p) is constant

Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.
(Include your Python code and output in the code box below.)

Ans:

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility (optional)
np.random.seed(42)

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean_value = np.mean(data)
std_value = np.std(data)

# Print results
print("Computed Mean:", mean_value)
print("Computed Standard Deviation:", std_value)

# Draw histogram
plt.figure()
plt.hist(data, bins=30)
plt.title("Histogram of Normal Distribution (Mean=50, SD=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Explanation

1000 random values were generated using np.random.normal() with:

Mean = 50

Standard Deviation = 5

Computed results:

Mean ‚âà 50.10

Standard Deviation ‚âà 4.89

The histogram shows a bell-shaped curve, confirming that the generated data follows a normal distribution.

```python
# Import required libraries
import numpy as np
import scipy.stats as stats

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
data = np.array(daily_sales)

# Sample size
n = len(data)

# Calculate sample mean
mean_sales = np.mean(data)

# Calculate sample standard deviation (ddof=1 for sample SD)
std_sales = np.std(data, ddof=1)

# Calculate Standard Error
standard_error = std_sales / np.sqrt(n)

# 95% confidence level
confidence_level = 0.95

# t-critical value (since sample size < 30 and population std unknown)
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

# Margin of Error
margin_of_error = t_critical * standard_error

# Confidence Interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

# Print results
print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```
